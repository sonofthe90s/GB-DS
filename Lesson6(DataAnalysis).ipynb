**В чём различие между зависимыми и независимыми выборками?**

Если одному случаю из выборки X соответствует один и только один случай из выборки Y, и наоборот,
то такие выборки называются зависимыми (например, муж и жена, либо измерение чего-нибудь в лаборатории до и после воздействия). 
В случае, если такая взаимосвязь между выборками отсутствует, 
то эти выборки считаются независимыми (например, футболист и фигурист, мужчина и женщина). 

Соответственно, зависимые выборки всегда имеют одинаковый объём, а объём независимых может отличаться.

**Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?**

Параметрический критерий используют, когда есть информация о характере распредления величины и достаточный объем выборки,
когда известны свойства выборки данных: тип распределения, среднее, дисперсия и т.д.

Непараметрический критерий используют, если нет информации о выборке или выборка очень малая (имеются данные низкого качества).
